```python
import pandas as pd
import re
from pathlib import Path

DATA_DIR = Path("./Regional_Data")
OUTPUT_MERGED_FILE = Path("./merged_dataset.csv")
OUTPUT_REPORT_FILE = Path("./missing_columns_report.txt")


COLUMN_MAPPING = {
    'age1': 'age',
    'fmtmetro3' : 'fmtmetro',
    'metro3' : 'metro',
    'istatus' : 'status'
}

def standardize_columns(columns: list) -> list:
    cleaned = []
    for col in columns:
        col = str(col).lower().strip()
        col = re.sub(r'[\s\-]+', '_', col)
        cleaned.append(col)
    return cleaned

def main():

    txt_files = list(DATA_DIR.glob("*.txt"))
    if not txt_files:
        return

    dataframes = []
    file_columns_map = {}

    for file in txt_files:
        try:

            df = pd.read_csv(file, sep=',', low_memory=False)
            df.columns = standardize_columns(df.columns)
            df.rename(columns=COLUMN_MAPPING, inplace=True)
            df['source_file'] = file.name 
            dataframes.append(df)
            file_columns_map[file.name] = set(df.columns) 
            
        except Exception as e:
            print("f")

    if dataframes:
      
        merged_df = pd.concat(dataframes, axis=0, ignore_index=True)

        merged_df.to_csv(OUTPUT_MERGED_FILE, index=False)


if __name__ == "__main__":
    main()
```
